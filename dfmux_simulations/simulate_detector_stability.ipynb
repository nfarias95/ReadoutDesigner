```python
# Plot different loop gains
loop_gains = np.linspace(5, 20, 4)
Rss = np.linspace(0, 0.5, 5)

R_TES_min_stable = np.zeros( (len(loop_gains) ,len(Rss))  )

for lg, loop_gain in enumerate(loop_gains):
    for rs, Rs in enumerate(Rss):
        R_TES_min_stable[lg, rs] = Rs * (loop_gain-1)/(loop_gain+1)

    plt.figure(1, figsize=FIG_SIZE)
    plt.plot(Rss, R_TES_min_stable[lg, :], label=str(loop_gain))
    plt.xlabel("Stray impedance Rs [Ohms]")
    plt.ylabel("Minimum stable TES resistance")
    plt.legend(title="Loop gain")
    plt.grid(True)


```
